```python
import pandas as pd
import altair as alt

df= pd.read_csv("/content/SNBP_2023_database.csv")
```

```python
df
```

```python
selection = alt.selection_interval()

alt.Chart(df).mark_bar().encode(
    x='Regiones ',
    y='Bibliotecas del SNBP',
    color=alt.condition(selection, 'Regiones ', alt.value('lightgray')),
    tooltip=['Regiones ', 'Bibliotecas del SNBP']
).add_selection(
    selection
).properties(
    width=600,
    height=400
).interactive()
```

```python
import altair as alt
import pandas as pd

df_melted = pd.melt(df2, id_vars=[], value_vars=['2021', '2022', '2023'],
                    var_name='Año', value_name='Total Servicios de la Red SNBP')


selection = alt.selection_interval()

alt.Chart(df_melted).mark_bar().encode(
    x=alt.X('Año:O', title='Año'),
    y=alt.Y('Total Servicios de la Red SNBP:Q', title='Total Servicios'),
    color=alt.condition(selection, 'Año:N', alt.value('lightgray')),
    tooltip=['Año', 'Total Servicios de la Red SNBP']
).add_selection(
    selection
).properties(
    title='Total Servicios de la Red SNBP por Año',
    width=600,
    height=400
).interactive()
```
