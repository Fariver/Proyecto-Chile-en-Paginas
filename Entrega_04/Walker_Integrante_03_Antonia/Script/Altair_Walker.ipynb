```python
import pandas as pd
import altair as alt

df= pd.read_csv("/biblio.csv")
```

```python
df
```

```python
import pandas as pd
import altair as alt
chart2 = alt.Chart(df).mark_bar().encode(
    x='Categoría',
    y='2019',
    color='Categoría',
    tooltip=['2019', 'Categoría']
).properties(
    width=600,
    height=400
).interactive()
display(chart2)
```
